## Boston house price dataset

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)
from sklearn.datasets import fetch_openml
boston = fetch_openml(name="boston", version=1, as_frame=True)

```

```python
print(boston.DESCR)
```

```python
print(boston.data)
```

```python
print(boston.target)
```

```python
print(boston.feature_names)
```

```python
dataset=pd.DataFrame(boston.data)
```

```python
dataset.head()
```

```python
dataset['Price']=boston.target
```

```python
dataset.head()
```

```python
dataset.info()
```

### sumarizing data sets

```python
dataset.describe()
```

## check the missing value

```python
dataset.isnull().sum()
```

### exploratory data analysis

```python
dataset.corr()
```

```python
plt.clf() 
plt.scatter(dataset['CRIM'],dataset['Price'])
plt.xlabel("Crime Rate")
plt.ylabel("Price")
plt.show()
```

```python
import seaborn as sns
sns.regplot(x="RM",y="Price",data=dataset)
plt.show()
```

### independent and dependent features

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

### Train Test Spli 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=40)
```

```python
X_train
```

```python
## standarise the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
### Model training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train , Y_train)
```

```python
## print the coefficient and the intercept 
```

```python
print(regression.coef_)
```

```python
 print(regression.intercept_)
```

```python
## on which parameter model has been trained 
regression.get_params()
```

```python
### prediction with test data
```

```python
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
### plot a scatterplot fro prediction 
```

```python
plt.scatter(Y_test,reg_pred)
```

```python
plt.show()
```

```python
## residuals 
```

```python
residuals=Y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind="kde")
```

```python
plt.show()
```

```python
### Scatter plot wrt prediction and residuals
```

```python
## uniform distributiom
plt.scatter(reg_pred,residuals)
```

```python
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(Y_test,reg_pred))
print(mean_squared_error(Y_test,reg_pred))
print(np.sqrt(mean_squared_error(Y_test,reg_pred)))
```

```python
### R square and adjusted R square

from sklearn.metrics import r2_score
score=r2_score(Y_test,reg_pred)
print(score)
```

### new data prediction
